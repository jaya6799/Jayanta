````python
## Solution

maximum = int(input(" Please Enter the Maximum Value : "))
minimum= int(input("Please Enter the Minimum Value: "))

while minimum <= maximum:
    if(minimum % 2 != 0):
        minimum += 1
    print(minimum)
    minimum += 2
````
